```python
import os
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, mutual_info_regression
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.linear_model import Ridge
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# =========================
# Constants
# =========================
RANDOM_STATE = 42
CSV_PATH = 'scenario3_all_variable_data_50000.csv'
FIRST_N_MINUTES = 20
SAMPLE_INTERVAL_MIN = 5
STEPS = int(FIRST_N_MINUTES / SAMPLE_INTERVAL_MIN)
TRAIN_CELLS = 30000

# =========================
# Helper Functions
# =========================
def find_columns(df):
    possible_id = ['storm_id', 'cell_id', 'cellID', 'ID', 'stormID', 'cellId']
    possible_time = ['timestamp', 'time', 'datetime', 'date_time', 'ts', 'timestamp_utc']
    id_col = next((c for c in possible_id if c in df.columns), None)
    time_col = next((c for c in possible_time if c in df.columns), None)
    return id_col, time_col

def load_data(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f"CSV not found at {path}.")
    df = pd.read_csv(path)
    id_col, time_col = find_columns(df)
    if time_col is None:
        df['__time_index__'] = df.groupby(id_col).cumcount()
        time_col = '__time_index__'
    else:
        try:
            df[time_col] = pd.to_datetime(df[time_col])
        except Exception:
            pass
    return df, id_col, time_col

def extract_first_n_features(df, id_col, time_col, steps=STEPS):
    feature_rows = []
    for sid, sub in df.groupby(id_col, sort=False):
        sub = sub.sort_values(time_col).head(steps)
        feat = {'storm_id': sid}
        for col in ['intensity_dbz', 'rainfall_mm_per_hr', 'intensity_change_rate']:
            if col in sub.columns:
                vals = sub[col].values
                feat[f'{col}_t0'] = vals[0]
                feat[f'{col}_tlast'] = vals[-1]
                feat[f'{col}_delta'] = vals[-1] - vals[0]
                feat[f'{col}_mean'] = np.nanmean(vals)
                feat[f'{col}_std'] = np.nanstd(vals)
                if len(vals) > 1:
                    x = np.arange(len(vals))
                    A = np.vstack([x, np.ones(len(x))]).T
                    m, _ = np.linalg.lstsq(A, vals, rcond=None)[0]
                    feat[f'{col}_slope'] = m
                else:
                    feat[f'{col}_slope'] = 0.0
                feat[f'{col}_lag1'] = sub[col].shift(1).iloc[1] if len(sub) > 1 else np.nan
        if 'x_position' in sub.columns and 'y_position' in sub.columns:
            feat['speed'] = np.sqrt((sub['x_position'].diff()**2 + sub['y_position'].diff()**2).mean()) / (SAMPLE_INTERVAL_MIN / 60) if len(sub) > 1 else 0.0
            feat['direction'] = np.arctan2(sub['y_position'].diff().mean(), sub['x_position'].diff().mean()) if len(sub) > 1 else 0.0
        if 'intensity_dbz' in sub.columns and 'size_pixels' in sub.columns:
            feat['intensity_size_interaction'] = sub['intensity_dbz'].mean() * sub['size_pixels'].mean()
        if 'time_since_formation_hours' in sub.columns and 'lifetime_hours' in sub.columns:
            feat['time_progress_ratio'] = sub['time_since_formation_hours'].mean() / sub['lifetime_hours'].iloc[0]
        feature_rows.append(feat)
    return pd.DataFrame(feature_rows).set_index('storm_id').fillna(0)

def compute_targets(df, id_col, time_col):
    rows = []
    for sid, sub in df.groupby(id_col, sort=False):
        sub = sub.sort_values(time_col)
        lifetime_h = sub['lifetime_hours'].iloc[0]
        peak_rain = sub['rainfall_mm_per_hr'].max()
        total_rain = (sub['rainfall_mm_per_hr'] * (SAMPLE_INTERVAL_MIN / 60.0)).sum()
        rows.append({
            'storm_id': sid,
            'lifetime_hours_true': lifetime_h,
            'peak_rainfall_mm_per_hr_true': peak_rain,
            'total_rainfall_mm_true': total_rain
        })
    return pd.DataFrame(rows).set_index('storm_id')

def regression_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    r2 = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / np.maximum(np.abs(y_true), 1e-6))) * 100
    return {'MAE': mae, 'RMSE': rmse, 'R2': r2, 'MAPE%': mape}

# =========================
# Load & process data
# =========================
df, id_col, time_col = load_data(CSV_PATH)
feats = extract_first_n_features(df, id_col, time_col, steps=STEPS)
targets = compute_targets(df, id_col, time_col)
data = feats.join(targets, how='inner').dropna()

# Fixed train/test split
train_ids, test_ids = train_test_split(
    list(data.index),
    train_size=TRAIN_CELLS,
    random_state=RANDOM_STATE
)

X = data.drop(columns=['lifetime_hours_true', 'peak_rainfall_mm_per_hr_true', 'total_rainfall_mm_true'])
y_life = data['lifetime_hours_true']
y_peak = data['peak_rainfall_mm_per_hr_true']
y_total = data['total_rainfall_mm_true']

X_train, X_test = X.loc[train_ids], X.loc[test_ids]
y_train_life, y_test_life = y_life.loc[train_ids], y_life.loc[test_ids]
y_train_peak, y_test_peak = y_peak.loc[train_ids], y_peak.loc[test_ids]
y_train_total, y_test_total = y_total.loc[train_ids], y_total.loc[test_ids]

# =========================
# Model Setup
# =========================
selector = SelectKBest(score_func=mutual_info_regression, k=min(10, X_train.shape[1]))
scaler = StandardScaler()
models = {
    'GB': GradientBoostingRegressor(random_state=RANDOM_STATE),
    'RF': RandomForestRegressor(n_estimators=100, random_state=RANDOM_STATE, n_jobs=-1),
    'Ridge': Ridge(random_state=RANDOM_STATE),
    'XGB': XGBRegressor(random_state=RANDOM_STATE)
}
param_grid = {
    'GB': {'model__n_estimators': [100, 200], 'model__learning_rate': [0.01, 0.1], 'model__max_depth': [3, 5]},
    'XGB': {'model__n_estimators': [100, 200], 'model__learning_rate': [0.01, 0.1], 'model__max_depth': [3, 5]},
    'RF': {'model__n_estimators': [100, 200], 'model__max_depth': [None, 10]},
    'Ridge': {'model__alpha': [0.1, 1.0, 10.0]}
}

# =========================
# Run Models
# =========================
results = []
for name, model in models.items():
    for target_name, y_train, y_test in [
        ('lifetime', y_train_life, y_test_life),
        ('peak', y_train_peak, y_test_peak),
        ('total', y_train_total, y_test_total)
    ]:
        pipe = Pipeline([('select', selector), ('scale', scaler), ('model', model)])
        grid = GridSearchCV(pipe, param_grid[name], cv=5, scoring='r2', n_jobs=-1)
        grid.fit(X_train, y_train)
        print(f"Best parameters for {name} - {target_name}: {grid.best_params_}")
        
        pred = grid.predict(X_test)
        metrics = regression_metrics(y_test.values, pred)
        print(f"{name} - {target_name}: {metrics}")
        results.append({'model': name, 'target': target_name, **metrics})
        
        cv_scores = cross_val_score(grid.best_estimator_, X_train, y_train, cv=5, scoring='r2')
        print(f"Cross-validated R2 for {name} - {target_name}: {cv_scores.mean():.4f} Â± {cv_scores.std():.4f}")
        
        if hasattr(grid.best_estimator_.named_steps['model'], 'feature_importances_'):
            importances = grid.best_estimator_.named_steps['model'].feature_importances_
            selected_features = X_train.columns[grid.best_estimator_.named_steps['select'].get_support()]
            print(pd.DataFrame({'Feature': selected_features, 'Importance': importances}).sort_values('Importance', ascending=False))
        
        plt.figure(figsize=(8, 6))
        residuals = y_test - pred
        plt.scatter(y_test, residuals, alpha=0.5)
        plt.axhline(0, color='red', linestyle='--')
        plt.xlabel(f'Actual {target_name}')
        plt.ylabel('Residuals')
        plt.title(f'Residual Plot for {name} - {target_name}')
        plt.show()

results_df = pd.DataFrame(results)
print("Final Results:")
print(results_df)
```

```python
import os
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.feature_selection import SelectKBest, mutual_info_regression
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.linear_model import Ridge
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# =========================
# Constants
# =========================
RANDOM_STATE = 42
CSV_PATH = 'scenario3_all_variable_data_50000.csv'
FIRST_N_MINUTES = 20
SAMPLE_INTERVAL_MIN = 5
STEPS = int(FIRST_N_MINUTES / SAMPLE_INTERVAL_MIN)
TRAIN_CELLS = 30000

# =========================
# Helper Functions
# =========================
def find_columns(df):
    possible_id = ['storm_id', 'cell_id', 'cellID', 'ID', 'stormID', 'cellId']
    possible_time = ['timestamp', 'time', 'datetime', 'date_time', 'ts', 'timestamp_utc']
    id_col = next((c for c in possible_id if c in df.columns), None)
    time_col = next((c for c in possible_time if c in df.columns), None)
    return id_col, time_col

def load_data(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f"CSV not found at {path}.")
    df = pd.read_csv(path)
    id_col, time_col = find_columns(df)
    if time_col is None:
        df['__time_index__'] = df.groupby(id_col).cumcount()
        time_col = '__time_index__'
    else:
        try:
            df[time_col] = pd.to_datetime(df[time_col])
        except Exception:
            pass
    return df, id_col, time_col

def extract_first_n_features(df, id_col, time_col, steps=STEPS):
    feature_rows = []
    for sid, sub in df.groupby(id_col, sort=False):
        sub = sub.sort_values(time_col).head(steps)
        feat = {'storm_id': sid}
        for col in ['intensity_dbz', 'rainfall_mm_per_hr', 'intensity_change_rate']:
            if col in sub.columns:
                vals = sub[col].values
                feat[f'{col}_t0'] = vals[0]
                feat[f'{col}_tlast'] = vals[-1]
                feat[f'{col}_delta'] = vals[-1] - vals[0]
                feat[f'{col}_mean'] = np.nanmean(vals)
                feat[f'{col}_std'] = np.nanstd(vals)
                if len(vals) > 1:
                    x = np.arange(len(vals))
                    A = np.vstack([x, np.ones(len(x))]).T
                    m, _ = np.linalg.lstsq(A, vals, rcond=None)[0]
                    feat[f'{col}_slope'] = m
                else:
                    feat[f'{col}_slope'] = 0.0
                feat[f'{col}_lag1'] = sub[col].shift(1).iloc[1] if len(sub) > 1 else np.nan
                feat[f'{col}_min'] = np.nanmin(vals)
                feat[f'{col}_max'] = np.nanmax(vals)
                feat[f'{col}_range'] = np.nanmax(vals) - np.nanmin(vals)
                feat[f'{col}_cv'] = (np.nanstd(vals) / (np.nanmean(vals) + 1e-6))
        if 'x_position' in sub.columns and 'y_position' in sub.columns:
            feat['speed'] = np.sqrt((sub['x_position'].diff()**2 + sub['y_position'].diff()**2).mean()) / (SAMPLE_INTERVAL_MIN / 60) if len(sub) > 1 else 0.0
            feat['direction'] = np.arctan2(sub['y_position'].diff().mean(), sub['x_position'].diff().mean()) if len(sub) > 1 else 0.0
        if 'intensity_dbz' in sub.columns and 'size_pixels' in sub.columns:
            feat['intensity_size_interaction'] = sub['intensity_dbz'].mean() * sub['size_pixels'].mean()
        if 'time_since_formation_hours' in sub.columns and 'lifetime_hours' in sub.columns:
            feat['time_progress_ratio'] = sub['time_since_formation_hours'].mean() / (sub['lifetime_hours'].iloc[0] + 1e-6)
        feature_rows.append(feat)
    return pd.DataFrame(feature_rows).set_index('storm_id').fillna(0)

def compute_targets(df, id_col, time_col):
    rows = []
    for sid, sub in df.groupby(id_col, sort=False):
        sub = sub.sort_values(time_col)
        lifetime_h = sub['lifetime_hours'].iloc[0]
        peak_rain = sub['rainfall_mm_per_hr'].max()
        total_rain = (sub['rainfall_mm_per_hr'] * (SAMPLE_INTERVAL_MIN / 60.0)).sum()
        rows.append({
            'storm_id': sid,
            'lifetime_hours_true': lifetime_h,
            'peak_rainfall_mm_per_hr_true': peak_rain,
            'total_rainfall_mm_true': total_rain
        })
    return pd.DataFrame(rows).set_index('storm_id')

def regression_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    r2 = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / np.maximum(np.abs(y_true), 1e-6))) * 100
    return {'MAE': mae, 'RMSE': rmse, 'R2': r2, 'MAPE%': mape}

# =========================
# Load & process data
# =========================
df, id_col, time_col = load_data(CSV_PATH)
feats = extract_first_n_features(df, id_col, time_col, steps=STEPS)
targets = compute_targets(df, id_col, time_col)
data = feats.join(targets, how='inner').dropna()

train_ids, test_ids = train_test_split(
    list(data.index),
    train_size=TRAIN_CELLS,
    random_state=RANDOM_STATE
)

X = data.drop(columns=['lifetime_hours_true', 'peak_rainfall_mm_per_hr_true', 'total_rainfall_mm_true'])
y_life = data['lifetime_hours_true']
y_peak = data['peak_rainfall_mm_per_hr_true']
y_total = data['total_rainfall_mm_true']

X_train, X_test = X.loc[train_ids], X.loc[test_ids]
y_train_life, y_test_life = y_life.loc[train_ids], y_life.loc[test_ids]
y_train_peak, y_test_peak = y_peak.loc[train_ids], y_peak.loc[test_ids]
y_train_total, y_test_total = y_total.loc[train_ids], y_total.loc[test_ids]

# =========================
# Model Setup
# =========================
selector = SelectKBest(score_func=mutual_info_regression, k=min(15, X_train.shape[1]))
scaler = StandardScaler()

models = {
    'GB': GradientBoostingRegressor(random_state=RANDOM_STATE),
    'RF': RandomForestRegressor(random_state=RANDOM_STATE, n_jobs=-1),
    'Ridge': Ridge(random_state=RANDOM_STATE),
    'XGB': XGBRegressor(random_state=RANDOM_STATE, tree_method='hist', enable_categorical=False, n_jobs=-1)
}

# Reduced param grids for speed
param_grid = {
    'GB': {
        'model__n_estimators': [200],
        'model__learning_rate': [0.05],
        'model__max_depth': [3]
    },
    'XGB': {
        'model__n_estimators': [200],
        'model__learning_rate': [0.05],
        'model__max_depth': [3]
    },
    'RF': {
        'model__n_estimators': [200],
        'model__max_depth': [None]
    },
    'Ridge': {
        'model__alpha': [1.0]
    }
}

# =========================
# Run Models
# =========================
results = []
for name, model in models.items():
    for target_name, y_train, y_test in [
        ('lifetime', y_train_life, y_test_life),
        ('peak', y_train_peak, y_test_peak),
        ('total', y_train_total, y_test_total)
    ]:
        pipe = Pipeline([
            ('select', selector),
            ('poly', PolynomialFeatures(degree=2, include_bias=False)),
            ('scale', scaler),
            ('model', model)
        ])
        grid = GridSearchCV(pipe, param_grid[name], cv=3, scoring='r2', n_jobs=-1)  # Reduced CV folds
        grid.fit(X_train, y_train)
        print(f"Best parameters for {name} - {target_name}: {grid.best_params_}")

        pred = grid.predict(X_test)
        metrics = regression_metrics(y_test.values, pred)
        print(f"{name} - {target_name}: {metrics}")
        results.append({'model': name, 'target': target_name, **metrics})

        # Use CV results from grid to avoid extra CV
        best_cv_mean = grid.cv_results_['mean_test_score'][grid.best_index_]
        best_cv_std = grid.cv_results_['std_test_score'][grid.best_index_]
        print(f"CV R2 for {name} - {target_name}: {best_cv_mean:.4f} Â± {best_cv_std:.4f}")

        # Feature importance
        if hasattr(grid.best_estimator_.named_steps['model'], 'feature_importances_'):
            selected_features = grid.best_estimator_.named_steps['select'].get_support(indices=True)
            selected_names = X_train.columns[selected_features]
            poly = grid.best_estimator_.named_steps['poly']
            feature_names = poly.get_feature_names_out(selected_names)
            importances = grid.best_estimator_.named_steps['model'].feature_importances_
            print(pd.DataFrame({'Feature': feature_names, 'Importance': importances})
                  .sort_values('Importance', ascending=False))

        # Residual Plot
        plt.figure(figsize=(8, 6))
        residuals = y_test - pred
        plt.scatter(y_test, residuals, alpha=0.5)
        plt.axhline(0, color='red', linestyle='--')
        plt.xlabel(f'Actual {target_name}')
        plt.ylabel('Residuals')
        plt.title(f'Residual Plot for {name} - {target_name}')
        plt.show()

results_df = pd.DataFrame(results)
print("Final Results:")
print(results_df)
```
